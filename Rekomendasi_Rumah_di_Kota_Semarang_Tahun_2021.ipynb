```python
plt.figure(figsize=(50,10))
p = sns.countplot(x="Harga", data=df)
```

**Insight:** Kecenderungan harga rumah lebih banyak dijual pada harga Rp600.000.000,00, diikuti pada harga Rp2.950.000.000,00,.

```python
CT = pd.crosstab(index=df["Harga"], columns=df["LT (m2)"])
p = CT.plot(kind="bar", figsize=(20,10), stacked=True)
```

**Insight:** Kecendurungan harga rumah Rp600.000.000,00, dengan luas tanah 127 $m^2$, diikuti harga rumah Rp2.950.000.000,00, dengan luas tanah 468 $m^2$.

**Rekomendasi:** Untuk pembeli yang mencari rumah dengan luas tanah yang luas

- Harga rumah di Kota Semarang cenderung meningkat seiring dengan luas tanahnya. Oleh karena itu, jika Anda mencari rumah dengan luas tanah yang luas, Anda harus siap untuk mengeluarkan biaya yang lebih besar.
- Harga rumah di Kota Semarang dengan luas tanah 468 m2 adalah sekitar Rp2.950.000.000,00. Harga ini cukup tinggi, tetapi cocok untuk Anda yang mencari rumah dengan luas tanah yang luas.
-  Untuk harga rumah disesuaikan dengan keinginan dan kebutuhan calon pembeli. Disarankan untuk harga rumah yang tinggi dengan tanah yang luas kepada keluarga besar, keluarga millenial, dan investor muda dan sukses.

```python
C = pd.crosstab(index=df["Listrik"], columns=df["LB (m2)"])
p = C.plot(kind="bar", figsize=(20,10), stacked=True)
```

**Insight:** Kecenderungan listrik dengan 2200 watt paling banyak terdapat pada rumah yang luas bangunannya 550 m2, 330 m2, 185 m2, 155 m2, 123 m2, 90 m2, 69 m2, 50 m2, 45 m2.

**Rekomendasi:**
Untuk pembeli yang mencari rumah dengan daya listrik yang besar

- Rumah dengan daya listrik 2200 watt paling banyak terdapat pada rumah dengan luas bangunan yang besar. Hal ini dikarenakan rumah dengan luas bangunan yang besar membutuhkan daya listrik yang lebih besar untuk memenuhi kebutuhan listriknya.
- Jika Anda mencari rumah dengan daya listrik yang besar, Anda dapat mempertimbangkan untuk membeli rumah dengan luas bangunan yang besar.
- Untuk listrik disesuaikan dengan kebutuhan calon pembeli, disarankan untuk calon pembeli rumah yang memilih bangunan luas dengan alat elektronik yang banyak, pilih rumah dengan kapasitas listrik yang tinggi.
